```python
import numpy as np

# 상관계수를 거리로 변환
distances = np.sqrt(2 * (1 - df_cor))
```

```python
mapping = {ticker: stock.get_market_ticker_name(ticker) for ticker in df_cor.columns}
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Nanum 폰트를 Matplotlib에 설정
plt.rc('font', family='NanumGothic')

# 그래프 생성
G = nx.Graph()

# 노드 추가
for node in df_cor.columns:
    G.add_node(node)

# 엣지 추가
for i in range(len(distances)):
    for j in range(i + 1, len(distances)):
        G.add_edge(distances.index[i], distances.columns[j], weight=distances.iloc[i, j])

# MST 생성
mst = nx.minimum_spanning_tree(G, weight='weight')

mst = nx.relabel_nodes(mst, mapping)

# 네트워크 레이아웃 설정
pos = nx.spring_layout(mst, scale=2)

# 시각화
plt.figure(figsize=(12, 8))
nx.draw(mst, pos, with_labels=True, node_size=20, node_color='skyblue', font_size=10, font_weight='bold', font_family='NanumGothic')
plt.title("Minimum Spanning Tree of Tickers")
plt.show()
```

```python
from matplotlib import cm

# 섹터별 색상 자동 설정
unique_sectors = meta_data['업종명'].unique()
colors = cm.get_cmap('tab20', len(unique_sectors))  # tab20 컬러맵 사용

sector_colors = {sector: colors(i) for i, sector in enumerate(unique_sectors)}

# 노드 색상 설정 함수
def get_node_colors(nodes):
    print(len(nodes))
    for i, node in enumerate(nodes):
        print(node)
        print(i, meta_data.loc[node, '업종명'])

    print("--------------")
    # return [sector_colors[meta_data.loc[node, '업종명']] for node in nodes]
```

```python
# MST 생성 및 시각화 함수
def create_mst_and_plot(corr_matrix, month, ax):
    # 상관계수를 거리로 변환
    distances = np.sqrt(2 * (1 - corr_matrix))

    # Nanum 폰트를 Matplotlib에 설정
    plt.rc('font', family='NanumGothic')

    # 그래프 생성
    G = nx.Graph()

    # 노드 추가
    print(corr_matrix.columns)

    for node in corr_matrix.columns:
        G.add_node(node)

    # 엣지 추가
    for i in range(len(distances)):
        for j in range(i + 1, len(distances)):
            G.add_edge(distances.index[i], distances.columns[j], weight=distances.iloc[i, j])

    # MST 생성
    mst = nx.minimum_spanning_tree(G, weight='weight')
    node_colors = get_node_colors(mst.nodes)

    mst = nx.relabel_nodes(mst, mapping)

    # 네트워크 레이아웃 설정
    pos = nx.spring_layout(mst, scale=2)

    # 노드 색상 설정

    # 시각화
    plt.figure(figsize=(12, 8))
    nx.draw(mst, pos, with_labels=True,ax=ax, node_size=80, node_color=node_colors, font_size=10, font_weight='bold', font_family='NanumGothic')
    ax.set_title(f"{month.strftime('%Y-%m')}")

    # 범례 추가
    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors(i), markersize=10) for i in range(len(unique_sectors))]
    labels = unique_sectors
    ax.legend(handles, labels, loc='best')

# 각 월별 상관 행렬 및 MST 시각화
fig, axes = plt.subplots(nrows=2, ncols=12, figsize=(100, 30))
axes = axes.flatten()  # 2D 배열을 1D 배열로 변환

for i, (month, monthly_df) in enumerate(monthly_groups):
    if i >= len(axes):
        break
    print(f"Month: {month}")
    monthly_df = monthly_df.T
    monthly_df.dropna(axis=0, how='any', inplace=True)  # NaN 값이 있는 열 제거
    corr_matrix = monthly_df.T.corr().fillna(0)  # NaN 값을 0으로 대체
    create_mst_and_plot(corr_matrix, month, axes[i])

plt.tight_layout()
plt.show()
```
